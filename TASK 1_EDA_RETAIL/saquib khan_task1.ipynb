```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
branch_sales=df.groupby('Branch')['Total'].sum().sort_values()

plt.figure(figsize=(8,5))
branch_sales.plot(kind= 'bar',color=['#FF6B6B','#4ECDC4','#45B7D1'])
plt.title('Branch wise Total sales',fontsize=15,fontweight='bold')
plt.xlabel('Branch')
plt.ylabel('Total sales')
plt.xticks(rotation=0)
plt.show()
```

```python
branch_sales=df.groupby('Branch')['Sales'].sum().sort_values()

plt.figure(figsize=(8,5))
branch_sales.plot(kind= 'bar',color=['#FF6B6B','#4ECDC4','#45B7D1'])
plt.title('Branch wise Sales',fontsize=15,fontweight='bold')
plt.xlabel('Branch')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
product_sales = df.groupby('Product line')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
product_sales.plot(kind='bar', color='skyblue')
plt.title('Product Line Wise Sales', fontsize=14, fontweight='bold')
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right')  # Naam tedhe ho jayenge taaki padh sako
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Payment'].value_counts().plot(kind='bar', color='orange')
plt.title('Payment Method usage')
plt.xlabel('Payment Type')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=0)
plt.show()
```
